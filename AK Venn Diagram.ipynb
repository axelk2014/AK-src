## Python code to generate Data Science Venn diagram

Generate a Venn diagram with the major components that make up "data science".

Make it look like an XKCD plot (https://xkcd.com/653/)

```python
import pandas as pd
import matplotlib.pyplot as plt; plt.rcdefaults()
from matplotlib_venn import venn3


# xkcd-ify matplotlib plots
plt.xkcd()

# create venn diagram
v=venn3(subsets = (6, 6, 2, 6,2, 2, 2), set_labels = ('', '', ''))

# add text and aesthetitcs
v.get_label_by_id('100').set_text('Computer\n Science')
v.get_label_by_id('111').set_size(12)
v.get_label_by_id('010').set_text('Mathematics')
v.get_label_by_id('111').set_size(12)
v.get_label_by_id('110').set_text('Machine\n Learning')
v.get_label_by_id('110').set_size(10)
v.get_label_by_id('001').set_text('Subject Matter\n Expertise')
v.get_label_by_id('111').set_size(12)
v.get_label_by_id('101').set_text('Kenny Logins*')
v.get_label_by_id('101').set_size(10)
v.get_label_by_id('101').set_url('https://www.youtube.com/watch?v=kyAn3fSs8_A')
v.get_label_by_id('011').set_text('Research')
v.get_label_by_id('011').set_size(10)
v.get_label_by_id('111').set_text('Data\n Science')
v.get_label_by_id('111').set_size(10)


#matplotlib.pyplot.savefig('datascience.svg',bbox_inches='tight')
plt.show()
```
